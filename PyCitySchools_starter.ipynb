```python
# Dependencies and Setup
import pandas as pd
import os
import csv

# File to Load (Remember to Change These)
# Read School and Student Data File and store into Pandas DataFrames

#School Data
school_data_csv = "Resources/schools_complete.csv"
school_data = pd.read_csv(school_data_csv)
school_data.head()

```

```python
#Student Data
student_data_csv = "Resources/students_complete.csv"
student_data = pd.read_csv(student_data_csv)
student_data.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = school_data["school_name"].nunique()
total_schools
```

```python
#Calculate the total number of students
total_students = int(student_data["Student ID"].nunique())
total_students
```

```python
#Calculate the total budget
total_budget = int(school_data["budget"].sum())
total_budget
```

```python
#Calculate the average math score
avg_math_score = round(student_data["math_score"].mean(), 1)
avg_math_score
```

```python
#Calculate the average reading score
avg_read_score = student_data["reading_score"].mean()
avg_read_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
#students with score above 70
passing_math = student_data[student_data["math_score"]>=70].count()["student_name"]

#calculating percentage of students passing math
percentage_passing_math = (passing_math/total_students)*100
percentage_passing_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
#students with score 70 or greater
passing_reading = student_data[student_data["reading_score"]>=70].count()["student_name"]

#calculating percentage of students passing reading
percentage_passing_reading = (passing_reading/total_students)*100
percentage_passing_reading
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = (student_data[(student_data["math_score"]>=70) & (student_data["reading_score"]>=70)]["school_name"].count()/total_students)*100
overall_passing
```

```python
#Create a dataframe to hold the above results

district_summary = pd.DataFrame({"Total Schools":[total_schools], 
                                 "Total Students": [total_students],
                                 "Total Budget": [total_budget], 
                                 "Average Math Score": [avg_math_score], 
                                 "Average Reading Score": [avg_read_score], 
                                 "% Passing Math": [percentage_passing_math], 
                                 "% Passing Reading": [percentage_passing_reading], 
                                 "% Overall Passing": [overall_passing]})
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#school type
school_type = school_data.set_index(["school_name"])["type"]
```

```python
#combining both files for summary on both files
combined_files = pd.merge(school_data, student_data, on="school_name")
combined_files.head()
```

```python
#total number of students per school
total_students_per_school = combined_files["school_name"].value_counts()

```

```python
#total school budget
budget_per_school = combined_files.groupby(["school_name"]).mean()["budget"]
budget_per_school
```

```python
#per student budget
budget_per_student = budget_per_school/total_students_per_school
budget_per_student
```

```python
#average math score
avg_math_scores = combined_files.groupby(["school_name"])["math_score"].mean()

```

```python
#average reading score
avg_read_scores = combined_files.groupby(["school_name"])["reading_score"].mean()

```

```python
# % passing math
passing_math = combined_files[(combined_files["math_score"]>=70)]
passing_math_per_school = passing_math.groupby(["school_name"]).count()["student_name"]
percentage_passing_math = (passing_math_per_school/total_students_per_school)*100
```

```python
# % Passing reading
passing_reading = combined_files[(combined_files["reading_score"]>=70)]
passing_reading_per_school = passing_reading.groupby(["school_name"]).count()["student_name"]
percentage_passing_reading = (passing_reading_per_school/total_students_per_school)*100
```

```python
# % Overall Passing
passing_overall = combined_files[(combined_files["math_score"]>=70) & (combined_files["reading_score"]>=70)]
passing_overall_per_school = passing_overall.groupby("school_name")["Student ID"].count()/total_students_per_school
percent_overall_passing = (passing_overall_per_school)*100

```

```python
#create school summary dataframe

school_summary = pd.DataFrame({
                               "School Type" : school_type,
                               "Total Students" : total_students_per_school,
                               "Total School Budget" : budget_per_school,
                               "Per Student Budget" : budget_per_student,
                               "Average Math Score" : avg_math_scores,
                               "Average Reading Score" : avg_read_scores,
                               "% Passing Math": percentage_passing_math,
                               "% Passing Reading" : percentage_passing_reading,
                               "Overall Passing Rate" : percent_overall_passing
    
                            })



school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# create the bins in which Data will be held
range_bins = [0, 584, 629, 644, 675]
# create group of categories
group_name = ["< $585", "$585-629", "$630-644", "$645-675"]

# putting data into categories
school_summary['Spending Ranges (Per Student)'] = pd.cut(school_summary['Per Student Budget'], bins = range_bins, labels = group_name)
# group the data set by spending ranges
by_spending_range = school_summary.groupby(['Spending Ranges (Per Student)'])

school_summary["Spending Ranges (Per Student)"]


```

```python
#calculating score and passing rate by Spending Ranges (Per Student)
mean_math_score = by_spending_range["Average Math Score"].mean()
mean_reading_score = by_spending_range["Average Reading Score"].mean()
mean_percent_math = by_spending_range["% Passing Math"].mean()
mean_percent_reading = by_spending_range["% Passing Reading"].mean()
overall_percent_pass = by_spending_range["Overall Passing Rate"].mean()

#df to store Scores by School Spending
scores_by_school_spending = pd.DataFrame({"Average Math Score":mean_math_score,
                                          "Average Reading Score":mean_reading_score,
                                          "% Passing Math":mean_percent_math,
                                          "% Passing Reading":mean_percent_reading,
                                          "Overall Passing Rate": overall_percent_pass
                                         })
# formatting it right
scores_by_school_spending['Average Math Score'] = scores_by_school_spending['Average Math Score'].map("{:,.2f}".format)
scores_by_school_spending['Average Reading Score'] = scores_by_school_spending['Average Reading Score'].map("{:,.2f}".format)
scores_by_school_spending['% Passing Math'] = scores_by_school_spending['% Passing Math'].map("{:.2f}".format)
scores_by_school_spending['% Passing Reading'] = scores_by_school_spending['% Passing Reading'].map("{:.2f}".format)
scores_by_school_spending['Overall Passing Rate'] = scores_by_school_spending['Overall Passing Rate'].map("{:.2f}".format)

scores_by_school_spending
```
